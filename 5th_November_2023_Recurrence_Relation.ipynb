````python
'''Q.3.). 
Given a recurrence relation, solve it using the recursive tree approach:
a. T(n) = 2T(n-1) +1
b. T(n) = 2T(n/2) + n
'''

# Solution :

'''
To solve the given recurrence relations using the recursive tree approach,
we'll visualize the recursion tree and then analyze its structure to find a pattern or a closed-form solution.

a. T(n) = 2T(n-1) + 1

Let's construct the recursion tree for this recurrence relation:

```
                            T(n)
                       /            \
                 T(n-1)            T(n-1)
              /       \          /       \
         T(n-2)      T(n-2)   T(n-2)   T(n-2)
          / \          / \       / \       / \
      T(n-3) T(n-3) T(n-3) T(n-3) T(n-3) T(n-3)
        .      .      .      .      .      .
        .      .      .      .      .      .
```

Each level of the tree has twice as many nodes as the previous level.
At each level, we add 1 to the total cost. So, the total cost at level k is 2^k. The tree's height is n,
so the total cost is:

T(n) = 1 + 2 + 2^2 + ... + 2^n = 2^{n+1} - 1

Therefore, T(n) = O(2^n).

b. T(n) = 2T(n/2) + n

Let's construct the recursion tree for this recurrence relation:

```
                            T(n)
                       /            \
                 T(n/2)            T(n/2)
              /       \          /       \
          T(n/4)    T(n/4)   T(n/4)    T(n/4)
           / \        / \       / \        / \
       T(n/8) T(n/8) T(n/8) T(n/8) T(n/8) T(n/8)
        .      .      .      .      .      .
        .      .      .      .      .      .
```

At each level, the cost increases by n because each subproblem has size n/2 and we add n at the root.
The tree's height is log n. So, the total cost is:

T(n) = O(n log n)

Therefore, T(n) = O(n log n).

These solutions provide insights into the growth rate of each recurrence relation based on their respective recursion tree structures.
'''
````
